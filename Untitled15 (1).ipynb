```python
# ========================================================
# Task A: Adjacency List Graph + Maze Visualization
# ========================================================
import matplotlib.pyplot as plt
import networkx as nx
import random
from collections import defaultdict

class AdjacencyListGraph:
    def __init__(self):
        self._adj = defaultdict(list)
        self._weights = {}

    def add_edge(self, u, v, w=1):
        self._adj[u].append((v, w))
        self._adj[v].append((u, w))
        self._weights[frozenset({u, v})] = w

    def vertices(self): return list(self._adj.keys())
    def edges(self): return [(tuple(e), w) for e, w in self._weights.items()]
    def neighbours(self, v): return self._adj[v]

    def print_graph(self):
        print("Adjacency List Representation:")
        for v in self._adj:
            print(f"  {v} -> {self._adj[v]}")
        print("Edges:", self.edges())

# ---- Maze Builder ----
def generate_grid_graph(n=4, m=4, weight_range=(1,3), seed=42):
    random.seed(seed)
    G = nx.grid_2d_graph(n, m)
    for (u, v) in G.edges():
        G.edges[u, v]['weight'] = random.randint(*weight_range)
    return G

def draw_maze(G, start=(0,0), n=4, m=4, title="Maze"):
    fig, ax = plt.subplots(figsize=(5,5))
    ax.set_title(title)
    ax.set_xlim(-0.5,m-0.5)
    ax.set_ylim(-0.5,n-0.5)
    ax.set_xticks(range(m))
    ax.set_yticks(range(n))
    ax.grid(True)
    for (u,v,d) in G.edges(data=True):
        w = d['weight']
        x1,y1=u; x2,y2=v
        color = {1:'lightblue',2:'blue',3:'darkblue'}[w]
        ax.plot([y1,y2],[x1,x2],color=color,linewidth=3)
    ax.scatter(start[1],start[0],c='red',marker='*',s=200,label="Start")
    ax.legend()
    plt.show()

# === Run Task A1 ===
print("=== Task A1 ===")
g1 = AdjacencyListGraph()
g1.add_edge("A","B",2); g1.add_edge("B","C",1); g1.add_edge("C","D",3)
g1.print_graph()
maze1 = generate_grid_graph(4,4,weight_range=(1,3),seed=1)
draw_maze(maze1,start=(0,0),title="Task A1 Maze")

# === Run Task A2 ===
print("\n=== Task A2 ===")
g2 = AdjacencyListGraph()
g2.add_edge("X","Y",5); g2.add_edge("Y","Z",2); g2.add_edge("X","Z",4)
g2.print_graph()
maze2 = generate_grid_graph(4,4,weight_range=(1,3),seed=5)
draw_maze(maze2,start=(0,0),title="Task A2 Maze")
```

```python
# ==============================
# Task B: Kruskalâ€™s MST + Image
# ==============================
import networkx as nx
import matplotlib.pyplot as plt
class UnionFind:
    def __init__(self, vertices): self.parent={v:v for v in vertices}
    def find(self,v):
        if self.parent[v]!=v: self.parent[v]=self.find(self.parent[v])
        return self.parent[v]
    def union(self,u,v):
        ru,rv=self.find(u),self.find(v)
        if ru==rv: return False
        self.parent[rv]=ru; return True

def kruskal_mst_graph(G):
    edges = sorted(G.edges(data=True), key=lambda e:e[2]['weight'])
    uf = UnionFind(G.nodes())
    mst_edges=[]
    for u,v,d in edges:
        if uf.union(u,v): mst_edges.append((u,v,d))
    MST = nx.Graph()
    MST.add_edges_from(mst_edges)
    return MST

def draw_mst(G, MST, title="Minimum Spanning Tree"):
    pos = {(i,j):(j,-i) for i,j in G.nodes()}
    plt.figure(figsize=(5,5))
    plt.title(title)
    nx.draw(G,pos,node_size=500,node_color="white",edge_color="lightgrey",with_labels=True)
    nx.draw_networkx_edges(MST,pos,width=2,edge_color="black")
    labels = nx.get_edge_attributes(MST,'weight')
    nx.draw_networkx_edge_labels(MST,pos,edge_labels=labels)
    plt.show()

# === Run Task B1 ===
print("=== Task B1 ===")
MST1 = kruskal_mst_graph(maze1)
draw_mst(maze1,MST1,title="Task B1 MST")

# === Run Task B2 ===
print("\n=== Task B2 ===")
MST2 = kruskal_mst_graph(maze2)
draw_mst(maze2,MST2,title="Task B2 MST")
```

```python
# ========================================================
# Task C: Prim vs Kruskal Timing
# ========================================================
import time, heapq

def prim_mst_graph(G, start=None):
    verts=list(G.nodes())
    if not verts: return nx.Graph()
    if start is None: start=verts[0]
    visited={start}
    edges=[(d['weight'],start,v) for v,d in G[start].items()]
    heapq.heapify(edges)
    MST=nx.Graph()
    while edges and len(visited)<len(verts):
        w,u,v=heapq.heappop(edges)
        if v in visited: continue
        visited.add(v)
        MST.add_edge(u,v,weight=w)
        for nbr,data in G[v].items():
            if nbr not in visited:
                heapq.heappush(edges,(data['weight'],v,nbr))
    return MST

def time_algorithms(n,density):
    G = nx.gnp_random_graph(n,density,seed=random.randint(1,100))
    for (u,v) in G.edges():
        G.edges[u,v]['weight'] = random.randint(1,10)
    t0=time.time(); prim_mst_graph(G); t1=time.time(); prim_t=t1-t0
    t0=time.time(); nx.minimum_spanning_tree(G); t1=time.time(); kruskal_t=t1-t0
    return prim_t,kruskal_t

sizes=[20,40,80]
densities=[0.2,0.5]

# === Run Task C1 ===
print("=== Task C1 ===")
for d in densities:
    prims=[]; kruskals=[]
    for n in sizes:
        p,k=time_algorithms(n,d)
        prims.append(p); kruskals.append(k)
    plt.figure()
    plt.plot(sizes,prims,marker="o",label="Prim")
    plt.plot(sizes,kruskals,marker="x",label="Kruskal")
    plt.xlabel("Nodes"); plt.ylabel("Time (s)")
    plt.title(f"Task C1 Runtime Comparison (density={d})")
    plt.legend(); plt.grid(True); plt.show()

# === Run Task C2 ===
print("\n=== Task C2 ===")
for d in [0.1,0.6]:
    prims=[]; kruskals=[]
    for n in sizes:
        p,k=time_algorithms(n,d)
        prims.append(p); kruskals.append(k)
    plt.figure()
    plt.plot(sizes,prims,marker="o",label="Prim")
    plt.plot(sizes,kruskals,marker="x",label="Kruskal")
    plt.xlabel("Nodes"); plt.ylabel("Time (s)")
    plt.title(f"Task C2 Runtime Comparison (density={d})")
    plt.legend(); plt.grid(True); plt.show()
```

```python
# ========================================================
# Task D: Clone-Based Exploration + Images (2 runs)
# ========================================================
from collections import defaultdict
import matplotlib.pyplot as plt
import networkx as nx # Import networkx

def build_tree(mst_edges):
    adj=defaultdict(list)
    for u,v,d in mst_edges.edges(data=True):
        adj[u].append((v,d['weight']))
        adj[v].append((u,d['weight']))
    return adj

def explore(MST,start):
    adj=build_tree(MST)
    explorers=[]
    def dfs(node,parent):
        cost=0
        for nbr,w in adj[node]:
            if nbr!=parent: cost=max(cost,dfs(nbr,node)+w)
        explorers.append((node,cost))
        return cost
    dfs(start,None)
    max_cost=max(c for _,c in explorers)
    return explorers,max_cost

def draw_exploration(MST, explorers, title="Exploration Plan"):
    pos={(i,j):(j,-i) for i,j in MST.nodes()}
    costs={n:c for n,c in explorers}
    node_colors=[costs[n] for n in MST.nodes()]

    # Create figure and axes explicitly
    fig, ax = plt.subplots(figsize=(5,5))

    # Draw nodes and edges, passing the axes
    nx.draw(MST,pos,with_labels=True,node_color=node_colors,
            cmap=plt.cm.viridis,node_size=800,font_size=8,edge_color="grey", ax=ax) # Pass ax
    labels=nx.get_edge_attributes(MST,'weight')
    nx.draw_networkx_edge_labels(MST,pos,edge_labels=labels,font_color="red", ax=ax) # Pass ax

    sm=plt.cm.ScalarMappable(cmap=plt.cm.viridis,
                             norm=plt.Normalize(vmin=min(node_colors), vmax=max(node_colors)))
    sm.set_array([])
    # Add colorbar, passing the axes
    plt.colorbar(sm, ax=ax, label="Explorer cost")
    plt.title(title)
    plt.show()

# === Run Task D1 ===
print("=== Task D1 ===")
# Ensure MST1 is defined from Task B (cell 612TwbY87fG1)
if 'MST1' in locals():
    explorers1,max_cost1=explore(MST1,(0,0))
    print("Explorers:",explorers1,"Max Cost:",max_cost1)
    draw_exploration(MST1, explorers1, title="Task D1 Exploration")
else:
    print("MST1 not defined. Please run Task B cell (612TwbY87fG1) first.")


# === Run Task D2 ===
print("\n=== Task D2 ===")
# Ensure MST2 is defined from Task B (cell 612TwbY87fG1)
if 'MST2' in locals():
    explorers2,max_cost2=explore(MST2,(0,0))
    print("Explorers:",explorers2,"Max Cost:",max_cost2)
    draw_exploration(MST2, explorers2, title="Task D2 Exploration")
else:
     print("MST2 not defined. Please run Task B cell (612TwbY87fG1) first.")
```
